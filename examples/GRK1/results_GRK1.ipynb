```python
import numpy as np

from matplotlib import pyplot as plt
```

```python
dat = np.loadtxt("GRK1.cp")

results = {
	'x': dat[:, 0],
	'Cp': dat[:, 1],
	'theta': dat[:, 2],
	'deltast': dat[:, 2] * dat[:, 3],
	'H': dat[:, 3],
	'N': dat[:, 4],
	'Cf': dat[:, 5],
	'y': dat[:, 6]
}
```

```python
plt.plot(results['x'], results['Cp'])

plt.title("$C_p$")
plt.xlabel("$x/c$")

plt.grid()

plt.show()
```

```python
plt.plot(results['x'], results['theta'])

plt.title(r"$\theta$")
plt.xlabel("$x/c$")

plt.grid()

plt.show()
```

```python
plt.plot(results['x'], results['deltast'])

plt.title(r"$\delta^*$")
plt.xlabel("$x/c$")

plt.grid()

plt.show()
```

```python
plt.plot(results['x'], results['H'])

plt.title("$H$")
plt.xlabel("$x/c$")

plt.grid()

plt.show()
```

```python
plt.plot(results['x'], results['N'])

plt.title("$N$")
plt.xlabel("$x/c$")

plt.grid()

plt.show()
```

```python
plt.plot(results['x'], results['Cf'])

plt.title("$C_f$")
plt.xlabel("$x/c$")

plt.ylim((0.0, 0.02))
plt.grid()

plt.show()
```

```python
dat = np.loadtxt("GRK1.plr")

polar = {
	'alpha': dat[:, 0],
	'CL': dat[:, 1],
	'CD': dat[:, 2],
	'Cm': dat[:, 3]
}

dat = np.loadtxt("GRK1_xfoil.plr", skiprows = 12)

polar_xfoil = {
	'alpha': dat[:, 0],
	'CL': dat[:, 1],
	'CD': dat[:, 2],
	'Cm': dat[:, 4]
}
```

```python
plt.plot(polar['alpha'], polar['CL'], linestyle = "-", label = "Current")
plt.plot(polar_xfoil['alpha'], polar_xfoil['CL'], linestyle = "--", label = "XFOIL", marker = '+')

plt.legend()
plt.grid()

plt.xlabel(r"$\alpha$")
plt.ylabel("$C_L$")

plt.show()
```

```python
plt.plot(polar['CL'], polar['CD'], linestyle = "-", label = "Current")
plt.plot(polar_xfoil['CL'], polar_xfoil['CD'], linestyle = "--", marker = "+", label = "XFOIL")

plt.legend()
plt.grid()

plt.xlabel("$C_L$")
plt.ylabel("$C_D$")

plt.show()
```
